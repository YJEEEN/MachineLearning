```python
from sklearn.datasets import load_digits
digits = load_digits()
samples = [0,10,20,30,1,11,21,31] # 이미지 번호
```

```python
digits # 숫자 이미지가 디지트화 되어 있는 데이터셋
```

```python
d = []
for i in range(8):
    d.append(digits.images[samples[i]]) #[0,10,20,30,1,11,21,31]
plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1,8,i+1) 
    plt.imshow(d[i], plt.cm.bone_r, interpolation='nearest')
    plt.title("img{}".format(i+1))
    plt.grid(False); plt.xticks([]); plt.yticks([])
plt.show()
```

### 2차원 벡터를 1차원 벡터로 펼치고자 한다.

```python
# d[0]이라함은 [0]번째 들어있는 2차원의 이미지 정보
# 2차원 (8,8) --> 1차원 벡터 (크기 64)
```

```python
v=[]
for i in range(8):
    v.append(d[i].reshape(64,1)) # 벡터화
    
# np.shape(v) 

plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1,8,i+1) 
    plt.imshow(v[i],aspect=0.2, interpolation='nearest')
    plt.title("img{}".format(i+1))
    plt.grid(False); plt.xticks([]); plt.yticks([])
plt.tight_layout(w_pad=3)
plt.show()
```

```python
from scipy import misc # misc에 컬러 이미지가 있다
img_rgb = misc.face()
img_rgb.shape # (가로, 세로, 채널 수)

plt.subplot(221) # 2행2열 1번째 위치
plt.imshow(img_rgb[:, :, 0],cmap=plt.cm.gray )
plt.axis("off")
plt.title("red img")

plt.subplot(222) # 2행2열 1번째 위치
plt.imshow(img_rgb[:, :, 1],cmap=plt.cm.gray )
plt.axis("off")
plt.title("green img")

plt.subplot(223) # 2행2열 1번째 위치
plt.imshow(img_rgb[:, :, 2],cmap=plt.cm.gray )
plt.axis("off")
plt.title("blue img")

plt.subplot(224) # 2행2열 1번째 위치
plt.imshow(img_rgb,cmap=plt.cm.gray )
plt.axis("off")
plt.title("rgb img")

plt.show()
```

### 행렬 연산의 특징

```python
x=np.array([[1,2],[3,4]])
y=np.array([[5,6],[7,8]])
z=np.array([[11,12],[13,14]])
print(x@y)
print(y@x)
print(x@(y+z))
print((x@y)+(x@z))
```
