```python
## 아이겐 밸류(고유값)
```

```python
고유벡터란?
: 행렬(A)이 작용하는 벡터(x)에 작용하는 힘의 방향
```

```python
선형 변환 행렬A의 고유값은 1,3이다 --> 행렬A를 1배, 3배하는 벡터
```

```python
x.T @ x = 공분산행렬
공분산행렬 --> 선형변환
```

```python
공분산 행렬의 고유벡터는?
: 데이터가 어떤 방향으로 분산되어 있는가를 의미 --> 고유값은 상관계수 즉, 각 축에 대한 공분산 값이다

고유값이 큰 순서대로 나열한다면,
가장 중요한 성분부터 정렬이 되는데 이것이 PCA
```

```python
차원의 저주
: 고차원으로 갈 수록 공간의 낭비가 크다 --> 차원을 줄여서 표현하자(100% 표현 못하지만)
    데이터 손실을 어느정도 받아들일 지에대한 기준이 필요하다(다만 분산이 큰 순서로)
```

```python
X=df[['calory', 'breakfast', 'lunch', 'dinner', 'exercise']]
X
Y=df[['body_shape']]
Y
```

```python
from sklearn.preprocessing import StandardScaler
x_std=StandardScaler().fit_transform(X)
x_std
```

```python
# 공분산행렬
# 피쳐가 컬럼에 있는데, row로 바꾸기 위해 변환
features=x_std.T
print(features)
cov_matrix=np.cov(features) # --> 공분산행렬 출력(5개의 피쳐에 대해 )
print(cov_matrix)
```

## np.linalg.eig(cov_matrix)
* 공분산 행렬로부터 아이겐벡터와 아이겐벨류 추출하는 함수
* 공분산행렬을 인자로 넣는다
* 아이겐밸류와 아이겐벡터를 반환한다
* 고유벡터는 해당 행렬이 벡터에 작용하는 힘의 방향
* 공분산의 고유벡터는? 데이터가 어느 방향으로 분산되어 있는지를 의미
* 고유값은 고유벡터에 해당하는 상관계수
* 공분산 행렬의 고유값은? 각 축에 해당하는 공분산 값
--> 즉, 고유값이 큰 순서대로 고유 벡터를 나열하면, 중요한 주성분부터 추출이 된다

```python
'''
A     X    rambda  X
2 1  0.7    3     0.7
1 2  0.7 =        0.7

A     X    rambda  X
2 1  0.7    1     0.7
1 2  -0.7 =      -0.7
'''
```

```python
# 공분산 행렬로부터 아이겐벡터와 아이겐벨류 추출
eig_vals,eig_vecs = np.linalg.eig(cov_matrix) # 공분산행렬을 인자로 넣는다 --> 아이겐밸루와 아이겐벡터를 반환한다
print("고유벡터:%s" % eig_vecs)
print("고유값:%s" % eig_vals)
```

```python
eig_vals[0]/sum(eig_vals) # --> pc1의 5차원 정보를 정사영하면, 73%의 정보를 표현할 수 있다는 의미
```

```python
projected_X=x_std.dot(eig_vecs.T[0])
projected_X # 일차원으로 차원 축소된 10행 5열 (원래 데이터)를 차원 축소한 결과
```

```python
### 2차원 공간에 시각화 해보자!
```

```python
res = pd.DataFrame(projected_X,columns=['PC1'])
res['yaxis']=0.0
res['label']=Y
res
```

```python
sns.lmplot('PC1','yaxis',data=res,hue='label') #5차원 행렬이던 데이터가 1차원으로 차원축소되어졌다(원 데이터 비해 73%의 정보를 담고)
```

### 표준화

```python
X_std=StandardScaler().fit_transform(X)
```

### 공분산

```python
mean_vec = np.mean(X_std,axis=0)
mean_vec
```

```python
cov_mat=(X_std-mean_vec).T.dot((X_std-mean_vec))/(X_std.shape[0]-1)
cov_mat
```

```python
np.cov(X_std.T)
```

```python
sns.heatmap(cov_mat,cmap='cubehelix',annot=True,square=True)
```

```python
eig_vals,eig_vecs=np.linalg.eig(cov_mat)
print("고유벡터:%s" % eig_vecs)
print("고유값:%s" % eig_vals)
```

```python
### 아이겐 벡터값을 내림차순으로 보는게 좋다
eig_pairs= [(eig_vals[i],eig_vecs[:,i])  for i in range(len(eig_vals))]
eig_pairs
```

```python
eig_pairs.sort(key=lambda x:x[0],reverse=True)
eig_pairs
```

```python
for i in eig_pairs:
    print(i[0])  # value들만 추출된다
    
tot =sum(eig_vals)

var_exp = [(i/tot)*100 for i in sorted(eig_vals,reverse=True)]
var_exp # 데이터를 어느정도 표현할 수 있는가에 대하 수치
```

```python
pca=PCA().fit(X_std)
pca
```

```python
pca.explained_variance_ratio_
```

### 이렇게 보고 데이터의 몇 %까지 정보를 나타낼건지 확인하고 차원을 선택하면 된다

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```
