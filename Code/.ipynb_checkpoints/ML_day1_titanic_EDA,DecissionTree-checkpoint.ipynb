### 분류
* 훈련 데이터의 특성 값/ 클래스 데이터로부터 학습한 뒤, 예측값을 얻는 것
* 데이터의 피쳐는 결과 값에 영향을 주는 질문(엔트로피)이 될 수 있다
* 주어진 데이터에 대해서만 잘 작동되고, 일반적이지 않은 경우 : 'over fitting' 되었다고 한다 
* 어떤 피쳐를 선택했을 때, 분류가 가장 잘 되는가 -->엔트로피 계산

### 결정 트리 구축 원칙
* 정보 획득량이 최대가 되는 특성(피쳐)를 찾는 것을 목표로 구축
1) 엔트로피 : 확률 변수의 불확실성을 수치로 나타낸다
2) IG(정보 획득량) : 분류 전 엔트로피 - 분류 후 엔트로피
                     즉, IG는 엔트로피(불확실성)의 감소량을 의미한다.
                     (엔트로피 1 일때, 분류가 안된 시점--> 가장 복잡하다)
---------------------------------------------------------------------------------
엔트로피가 작을수록, IG가 클수록 결정 트리

```python
# 계속해서 호출 할 바차트 함수를 만든다
def bar_chart(f):
    train[train['Survived']==1][f] # survived==1인 사람들의 sex를 출력
    survived = train[train['Survived']==1][f].value_counts() 
    dead = train[train['Survived']==0][f].value_counts()
    df = pd.DataFrame([survived, dead])
    df.index=['Survived','Dead']
    df.plot.bar(stacked=True, figsize=(8,4)) # default는 line이다 #stacked 옵션:하나로 쌓아서 표현
#     print(df)

    
bar_chart('Sex') #성별에 따른 생존자수를 막대그래프  
```

```python
bar_chart('Pclass')
```

```python
bar_chart('SibSp')
```

```python
train['Age_median']=train['Age'] # 중앙값
train['Age_mean']=train['Age'] # 평균
train['Age_mode']=train['Age'] # 최빈수

train['Age_median'] = train['Age_median'].fillna(train['Age_median'].median())
train['Age_mean'].fillna(train['Age_mean'].mean(), inplace=True)
train['Age_mode'].fillna(train['Age_mode'].mode())
train['Age_mode'].fillna(train['Age_mode'].mode()[0],inplace=True)
print(train['Age_mode'].mode()) # 0    24.0 --> 함수의 수행결과가 
# type(train['Age_mode'].mode())
print(train['Age_mode'].mean()) # 29.69911764705882 --> 수행결과는 스칼라 값이다
train.info()
```
