# 연관성 규칙

```python
문장=transaction
문장내 단어=item
--> 문서 요약에 활용
```

### 지지도

```python
전체 거래 중 x와y를 함께(동시에) 포함하는 거래가 어느 정도인가
```

### 신뢰도

```python
x를 구입한 사람이 y를 구매할 확률
아이템 집합이 있을때,
A아이템 집합에서 item을 구입한 사람이 B아이템 집합에서 item을 구입할 확률
```

### 리프트(향상도)

```python
x가 구입되었다면, y가 구매될 확률(신뢰도)
--------------------------------------  =  향상도
y의 일반적인 구매 확률(y구매/전체 구매)
```

```python
ex)
 lift(우유->빵) > 1이라면, 무슨 의미인가?
    : c(우유->빵)의 값이 빵 구매 확률이 높다.
        즉, 우유를 산 사람이 빵을 살 확률이 높다.
```

## Apriori Algorithm

```python
pruming(가지치기)
: 일정 수치 이상의 지지도를 가진 item 집합에 대해서만 연관성 규칙을 찾는 것
    p(a,b)는 a,b를 같이 구매할 확률, 
    a집합과 b집합의 초월집합(superset)이라고 함
    
***적절한 pruming을 선택하는 것이 중요한 issue이다***
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
sample=" Mulmi team, cheer-up"
```

```python
idx2char=list(set(sample))
char2idx={c:i for i,c in enumerate(idx2char)}
```

```python
char2idx
dic_size=len(char2idx)
hidden_size=num_classes=dic_size
num_classes # 최종출력 output size
hidden_size #output size
dic_size  #원핫인코딩크기

batch_size=1 # data 1개씩 덥데이트 하겠다
sequence_length=len(sample)-1
learning_rate=0.1
```

```python
# 해당 문자의 인덱스 번호를 리스트형태로 저장
sample_idx= [char2idx[c] for c in sample]
sample_idx
sample_idx[:-1]
xdata=[sample_idx[:-1]]
ydata=[sample_idx[1:]]
print(xdata)
print(ydata)
```

```python
tf.reset_default_graph()
x=tf.placeholder(tf.int32,[None,sequence_length])
y=tf.placeholder(tf.int32,[None,sequence_length])

x_one_hot=tf.one_hot(x,num_classes)

cell=tf.contrib.rnn.BasicLSTMCell(
num_units=hidden_size)
init_state=cell.zero_state(batch_size=1,dtype=tf.float32)
outputs,_states=tf.nn.dynamic_rnn(cell,x_one_hot,initial_state=init_state,dtype=tf.float32)
```

```python
xfc=tf.reshape(outputs,[-1,hidden_size])
# fully-connected에서 default가 'relu'이다 -->  None으로 설정하면 선형적으로 진행이된다
outputs=tf.contrib.layers.fully_connected(xfc,num_classes,activation_fn=None)
outputs=tf.reshape(outputs,[batch_size,sequence_length,num_classes])
outputs

weights=tf.ones([batch_size,sequence_length])
sequence_loss=tf.contrib.seq2seq.sequence_loss(
logits=outputs,targets=y,weights=weights)
loss=tf.reduce_mean(sequence_loss)
train=tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(loss)
prediction=tf.argmax(outputs,2)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(100):
        cv,_=sess.run([loss,train],
                     feed_dict={x:xdata,y:ydata})
        res=sess.run(prediction,feed_dict={x:xdata})
        res_str=[idx2char[c] for c in np.squeeze(res)]
        print(i,"loss:", cv, "prediction:","".join(res_str))
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
dataset=[['사과','치즈','생수'],
        ['생수','호두','치즈','고등어'],
        ['수박','사과','생수'],
        ['생수','호두','치즈','옥수수']]
```

```python
te=TransactionEncoder()
# transform()를 통해 종류별 boolean값을 return하는 형태로 변환!!
te_ary=te.fit(dataset).transform(dataset)
te_ary
```

```python
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
freq_item=apriori(df,min_support=0.5,use_colnames=True)
freq_item
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(freq_item,metric="confidence",min_threshold=0.3)
```

```python
association_rules(freq_item,metric="lift",min_threshold=0.7)
```
