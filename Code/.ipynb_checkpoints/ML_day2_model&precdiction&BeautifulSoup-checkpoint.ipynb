```python
# 디씨젼트리를 이용한 분류
import pandas as pd
import numpy as np
import urllib
import urllib.request as req
from bs4 import BeautifulSoup
```

```python
train = pd.read_csv("/Users/user/Downloads/titanic/train.csv")
train
test = pd.read_csv("/Users/user/Downloads/titanic/test.csv")
test
```

```python
train.describe() # 기술 통계가 구해진다
```

```python
# 평균 값으로 age를 채우고 별도로 저장하겠다
train['Age_mean'] = train['Age']
train['Age_mean'].fillna(train['Age_mean'].mean(),inplace=True) # inplace는 따로 assign하지 않아도 된다
test['Age_mean'] = test['Age']
test['Age_mean'].fillna(test['Age_mean'].mean(),inplace=True)
```

```python
train['Sex'].value_counts()
```

```python
train['Gender'] = (train['Sex']=='female')
test['Gender'] = (test['Sex']=='female')
```

```python
train['Gender']
# train['Sex']
```

```python
train['Embarked'].value_counts()
train['Embarked'].isnull().sum() # 2개가  NAN
test['Embarked'].isnull().sum() # 0개가  NAN
```

```python
# 원핫인코딩
# Embarked_S를 만들어서 Embarked == 'S'인지 확인하고 할당하고, 나머지에 대해서도 동일하게 수행

train['Embarked_S'] = train['Embarked']=='S'
train['Embarked_C'] = train['Embarked']=='C'
train['Embarked_Q'] = train['Embarked']=='Q'

# train['E']
```

```python
test['Embarked_S'] = test['Embarked']=='S'
test['Embarked_C'] = test['Embarked']=='C'
test['Embarked_Q'] = test['Embarked']=='Q'
```

```python
train['FamilySize'] = train['SibSp'] + train['Parch']+1 # +1(본인)까지 하면 --> 총 가족의 수 
train['FamilySize'].value_counts()  # --> 인덱스가 가족의 수
train['Family'] = train['FamilySize']

# 대가족(5이상) :L 중가족(2-4) :M 소가족(1):S 로 분류한다면,
train.loc[train['FamilySize']==1,'Family'] = 'S'# loc를 사용하여 가족수 열이 1과 같은 'Famliy'열을 참조하겠다
train.loc[ (1<train['FamilySize']) & (train['FamilySize']<5) ,'Family'] = 'M'
train.loc[train['FamilySize']>4,'Family'] = 'L'
```

```python
train['Family_S']=train['Family']=='S'
train['Family_M']=train['Family']=='M'
train['Family_L']=train['Family']=='L'

train[['Family_S','Family_M','Family_L','Family','FamilySize']]
train['Family'].value_counts()
```

```python
test['FamilySize'] = test['SibSp'] + test['Parch']+1 # +1(본인)까지 하면 --> 총 가족의 수 
test['FamilySize'].value_counts()  # --> 인덱스가 가족의 수
test['Family'] = test['FamilySize']
test['Family'] = test['FamilySize']

test.loc[train['FamilySize']==1,'Family'] = 'S'# loc를 사용하여 가족수 열이 1과 같은 'Famliy'열을 참조하겠다
test.loc[ (1< test['FamilySize']) & (test['FamilySize']<5) ,'Family'] = 'M'
test.loc[train['FamilySize']>4,'Family'] = 'L'
test[['FamilySize','Family']]

test['Family_S']=test['Family']=='S'
test['Family_M']=test['Family']=='M'
test['Family_L']=test['Family']=='L'
test[['Family_S','Family_M','Family_L','Family','FamilySize','Embarked_S','Embarked_C','Embarked_Q']]

```

```python
train['Pclass'].dtypes
```

```python
train['Pclass'] = train['Pclass'].astype('category')
train['Pclass'].dtypes
```

```python
fn=['Gender','Age_mean','Embarked_S', 'Embarked_C', 'Embarked_Q',
    'Family_S','Family_M','Family_L']
# train.info()
X_train=train[fn]
X_train.head()
y_label=train['Survived']
X_test=test[fn]
# X_train과 y_label을 이용하여 모델을 만들겠다
```

```python
# 모델링(디시전 트리)
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=3) # 디시전 트리 객체가 만들어짐 --> IG 를 알아서 구해준다
model
```

```python
model.fit(X_train,y_label)
# X_train.isnull().sum()
```

```python
from sklearn.model_selection import KFold, cross_val_score
kfold=KFold(n_splits=10,shuffle=True,random_state=2019) 
#n_split=3, shuffle=False, random_state=None, 
# random_State는 임의로 설정
kfold 
```

```python
score = cross_val_score(model, X_train, y_label, cv=kfold, scoring='accuracy') 
print(score)
round(np.mean(score)*100,2) # round(,2)로 소숫점 이하 둘째자리까지 출력
```

```python
prediction = model.predict(X_test) # survived 칼럼에 대한 값이므로 0,1 로 출력되고 이건 모델이 예측한 값이다
prediction.shape
```

```python
test['Survived']=prediction
test.columns
```

```python
submission = test[['PassengerId','Survived']]
submission.to_csv("submit.csv",index=False)
```

```python
# findall()을 이용한 여러 요소 추출
html = """
<html><body>
<ul>
<li><a href="http://naver.com">naver</a></li> 
<li><a href="http://daum.net">daum</a></li>
</ul>
</body></html>
"""

soup = BeautifulSoup(html,'html.parser')
soup.find_all("ul") # findall이라서 리스트 형태로 출력된다
soup.find_all("li") # 같은 태그명으로 묶인 모든 태그를 불러올 수 있다
links = soup.find_all("a")

for link in links :
#     print(link.attrs['href']) # attrs[추출하려는 속성이름]
    href = link.attrs['href']
#     print(link.string)
    text = link.string
    print(text,"==>",href)
```

```python
# 기상청으로 이동해서
# 1. 타이틀태그내용추출
# 2. wf태그 내용 추출
# 3. tmn,tmx 값 추출
```

```python
values={'stnId':'184'} # 제주도
params = urllib.parse.urlencode(values)
API ="https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
url = API+"?"+params
url

# jeju = pd.DataFrame()
WF=[]
TMN=[]
TMX=[]

data = urllib.request.urlopen(url).read() # --> xml문서로 봤던 내용 모두가 출력됨
text = data.decode('utf-8')
# print(text)

soup = BeautifulSoup(text,'html.parser')

titles = soup.find_all("title") 
for title in titles:
    print(title.string)
    
wfs = soup.find_all("wf")
for wf in wfs:
    print(wf.string)
#     jeju['wf'] = title.string
    WF.append(wf.string)
    
tmns = soup.find_all("tmn")
tmxs = soup.find_all("tmx")
for tmn in tmns:
    print(tmn.string)
#     jeju['tmn'] = title.string
    TMN.append(tmn.string)

for tmx in tmxs:
    print(tmx.string)
#     jeju['tmx'] = title.string
    TMX.append(tmx.string)
    
Jeju = pd.DataFrame(np.array([WF,TMN,TMX]).T,columns=["wf","tmn","tmx"])
Jeju
```

```python
# CSS selector : jQuery처럼 css선택자 사용가능 --> 특정 요소 추출하는 기능 제공
#같은 태그 중 범위를 지정하는 역할을 해준다  
# soup.select(<선택자>) : 여러 개의 요소 --> 태그 아래 태그까지 지정 가능
# soup.select_one(<선택자>) : 1개 요소
```

```python
html="""
<html><body>
<h1>역삼동</h1>
<div id="myid">
<h1>멀티캠퍼스</h1>
<ul class="items">
<li>강의실</il>
<li>엘레베이터</il>
<li>식당</il>
</ul>
</div></body></html>
"""

soup = BeautifulSoup(html,'html.parser')
print(soup.select_one("h1"))
print(soup.select_one("li"))
print(soup.select("li")) # findall처럼 전체가 출력됨
# h1태그가 역삼동과 멀캠 2가지가 있는데,
# 역삼동은 html 아래 소속, 멀티캠퍼스는 div태그 아래 소속되어 있다.
# div태그는 id가 주어져있어서 div#id이름 으로쓰면 멀티캠퍼스를 출력할 수 있다
print(soup.select_one("div#myid h1")) # id이름 뒤에 spacebar주고, 하위 태그명을 달면 그 태그가 출력된다
print(soup.select_one("div#myid > h1")) # >를 표기하여 출력할 수도 있다

soup.select("div#myid ul.items li") # class이름이 부여된 경우, '.'찍고 클래스명 spacebar 하위태그 로도 특정 태그를 출력할 수 있다
```
