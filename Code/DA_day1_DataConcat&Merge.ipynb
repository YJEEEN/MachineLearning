```python
# merge() : how(left, right, inner(default), outer)
# on(병합 기준 변수)
```

```python
dfleft = pd.DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']})


dfright = pd.DataFrame({'KEY': ['K2', 'K3', 'K4', 'K5'],
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']})
```

```python
dfleft
```

```python
dfright
```

```python
pd.merge(dfleft,dfright) # 공통적으로 가지고 있는 column을 찾은 후, 기본적으로 inner join(교집합)을 진행
#pd.merge(dfleft,dfright,on='KEY') 위에 명령어와 같은 말임
```

```python
pd.merge(dfleft,dfright,on='KEY', how = 'left') 
# left를 기준으로 병합되어지는 대상의 정보가 출력 (기준의 정보는 모두 출력됨)
```

```python
pd.merge(dfleft,dfright,on='KEY', how = 'inner') 
```

```python
pd.merge(dfleft,dfright,on='KEY', how = 'outer') 
```

```python
df2 = pd.DataFrame({'a1':[1,2,3,4,5],
                   'a2':[10,11,12,13,14]})
df2.ix[[1,3],['a2']] = np.nan
df2
```

```python
# a3 열 추가
# a3 열이 NaN이면, a1열의값으로 a3열을 채우고,
# a3 열이 NaN이 아니면, a2열의 값으로 a3를 채우고
# a3: 10,2,12,4,14
# df2 출력

df2['a3'] = np.where(pd.notnull(df2['a2'])==True, df2['a2'],df2['a1'])
df2
```
