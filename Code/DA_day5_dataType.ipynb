```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import re
```

```python
tips = sns.load_dataset("tips")
tips
```

### 형변환 함수01 .astype(데이터타입)

```python
tips['sex_str']=tips['sex'].astype(str)
tips.info()
```

```python
tips['total_bill']=tips['total_bill'].astype(str)
tips['total_bill']=tips['total_bill'].astype(float)
tips.info()
```

```python
tips_sub_miss = tips.head(10)
tips_sub_miss.loc[[1,3,5,7],'total_bill']='missing'
tips_sub_miss
tips_sub_miss.info()
tips_sub_miss['total_bill'].astype(float) # --> could not convert string to float: 'missing'
```

```python
pd.to_numeric(tips_sub_miss['total_bill'])
```

### 형변환 함수02 .to_numeric(errors='ignore/coerce/raise')

```python
pd.to_numeric(tips_sub_miss['total_bill'],errors='ignore')
pd.to_numeric(tips_sub_miss['total_bill'],errors='coerce') #--> 숫자로 변환되지 않는 데이터는 NaN으로 처리
# pd.to_numeric(tips_sub_miss['total_bill'],errors='raise') #--> 숫자로 변환되지 않는 데이터는 에러로 처리:default
```

```python
pd.to_numeric(tips_sub_miss['total_bill'],errors='coerce',downcast='float') # casting은 변환할 때 쓰는 용어
                                                                            # downcast은 float으로 다운시켜서 변환하라는 것
                                                                    #float64에서 float32로 다운캐스팅됨
```
