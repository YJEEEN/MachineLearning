# 2) 차원 축소를 수행한 후 분류 모델기 생성

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.columns
```

```python
X=df[['satisfaction_level','promotion_last_5years','Work_accident','time_spend_company']]
X
Y=df[['left']]
Y
```

```python
x_std=StandardScaler().fit_transform(X)
x_std
```

```python
features=x_std.T
print(features)
cov_matrix=np.cov(features) # --> 공분산행렬 출력(5개의 피쳐에 대해 )
print(cov_matrix)
```

```python
# 공분산 행렬로부터 아이겐벡터와 아이겐벨류 추출
eig_vals,eig_vecs = np.linalg.eig(cov_matrix) # 공분산행렬을 인자로 넣는다 --> 아이겐밸루와 아이겐벡터를 반환한다
print("고유벡터:%s" % eig_vecs)
print("고유값:%s" % eig_vals)
```
