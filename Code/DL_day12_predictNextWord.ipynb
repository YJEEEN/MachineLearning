```python
import numpy as np
```

```python
timesteps=10 # 시점, 문장을 구성하는 10개의 단어가 input으로 들어가기 위해 준비되어 있다.
# 주식 예측 모델에서 timestep=7이면, 7일간의 주식변동 패턴으로 오늘 주가를 예상하겠다
input_dim=4 #단어가4개인 벡터(코퍼스내의 단어의 갯수)
hidden_size=8 # 은닉상태의 크기(셀)
inputs=np.random.random((timesteps,input_dim))
inputs
```

### 초기 은닉 상태를 0으로 초기화

```python
hidden_state_t=np.zeros((hidden_size,))

#가중치 (은닉상태의크기*입력차원)
wx=np.random.random((hidden_size,input_dim)) #(8, 4)
#은닉상태 가중치 (은닉상태의크기*은닉상태의크기)
wh=np.random.random((hidden_size,hidden_size)) #(8, 8)
#b바이어스(은닉상태의크기)
b=np.random.random((hidden_size,)) #8

print(np.shape(wx))
```

### 메모리 셀 동작

```python
total_hidden_state=[]

for input_t in inputs:
#     print(input_t)
    output_t=np.tanh(np.dot(wx,input_t)+np.dot(wh,hidden_state_t)+b) # wx*input+wh*ht-1+b    
#     print(output_t)
    total_hidden_state.append(list(output_t))
#     print(total_hidden_state)
    hidden_state_t=output_t #다음 입력셀로 현재의 상태를 넘겨줌
```

```python
from keras.preprocessing.text import Tokenizer 
```

```python
t=Tokenizer()
t.fit_on_texts([text]) #단어에 인덱스 붙여줌
t.texts_to_sequences([text]) #인덱스만 단어 순서에 따라 출력
print(t.texts_to_sequences([text]))
print(t.word_index) #단어당 인덱스를 함께 출력
```

```python
vocab_size=len(t.word_index)+1 #단어 집합의 크기:13
# 입력:멀티캠퍼스 , 단어수2
# 출력:있는 교육생들이

for line in text.split("\n"):
    print(line)
    print(t.texts_to_sequences([line])[0])

    
sequences=list()
for line in text.split("\n"):
#     print(t.texts_to_sequences([line])[0])
    encoded=t.texts_to_sequences([line])[0]
    for i in range(1,len(encoded)):
        sequence=encoded[:i+1]
#         print(sequence)
        sequences.append(sequence)
print(len(sequences))
print(sequences)
        
print(max(len(s) for s in sequences))
max_len=max(len(s) for s in sequences)
```

```python
from keras.preprocessing.sequence import pad_sequences
sequences=pad_sequences(sequences,maxlen=max_len)
sequences
```

```python
x=sequences[:,:-1]
y=sequences[:,-1]
x
y
```

### one-hot encoding

```python
from keras.utils import to_categorical
```

```python
y=to_categorical(y,num_classes=vocab_size)
y
```

```python
from keras.layers import *
x
model=Sequential()
model.add(Embedding(vocab_size,10,input_length=max_len-1))
model.add(SimpleRNN(32))
model.add(Dense(vocab_size,activation='softmax'))
model.compile(loss='categorical_crossentropy',
             optimizer='adam',metrics=['accuracy'])
model.fit(x,y,epochs=300,verbose=2) 
```

```python
def word_gen(model, t, curr_word, num):
          #모델,토큰나이저,입력단어,개수
        ini_word=curr_word
        sent=''
        for _ in range(num):
            #print(t.word_index)
            encoded=t.texts_to_sequences([curr_word])[0] #[3]
            #print(encoded)
            encoded=pad_sequences([encoded],maxlen=5,padding='pre') # [[0 0 0 0 3]]
            #print(encoded)           
            res=model.predict_classes(encoded)
            #print(res) #4
            for word, index in t.word_index.items():
                if res==index: #동일한 단어를 찾았다면
                    break
            curr_word=curr_word+' '+word 
            sent=sent+' '+word            
        sent=ini_word+sent
        return sent      
print(word_gen(model, t, '멀캠', 10))    
#멀티캠퍼스에 라는 단어 뒤에 3개의 단어 예측
```

```python
headline=[]
headline.extend(list(df.headline.values))
headline

headline=[n for n in headline if n != 'Unknown']
len(headline)
# headline
#     print(n)
```

```python
from string import punctuation
def mypreprocessing(s):# 점 제거,소문자로 변환
    # 전처리된 문장을 return하겠다.
#     s.lower()
#     print(s)
#     print(punctuation)
#     return s#전처리된 문장
#     print(''.join(ch for ch in s if ch not in punctuation).lower())
    return ''.join(ch for ch in s if ch not in punctuation).lower()

text=[mypreprocessing(x) for x in headline] #변환되기 전 문장
print(len(text))
```

```python
# vocabulary 작성
t=Tokenizer()
t.fit_on_texts(text)
t.word_index
vocab_size=len(t.word_index)+1 # 단어 집합의 크기:3620개(중복되지 않는 단어 갯수)
vocab_size
```

```python
sequences=[]
for line in text: #1214개의 라인
    encoded=t.texts_to_sequences([line])[0]
    for i in range(1,len(encoded)):
        sequence=encoded[:i+1]
        sequences.append(sequence)
    print(sequences[:20])
```

```python
max_len_headline=max(len(s) for s in sequences )
```

```python
from keras.preprocessing.sequence import pad_sequences
sequences=pad_sequences(sequences,maxlen=max_len_headline)
sequences
```

```python
x=sequences[:,:-1]
y=sequences[:,-1]
x
y
```

```python
y=to_categorical(y,num_classes=vocab_size)
y
```

```python
from keras.layers import *
x
model=Sequential()
model.add(Embedding(vocab_size,10,input_length=max_len_headline-1))
model.add(SimpleRNN(32))
model.add(Dense(vocab_size,activation='softmax'))
model.compile(loss='categorical_crossentropy',
             optimizer='adam',metrics=['accuracy'])
model.fit(x,y,epochs=300,verbose=2) 
```

```python
def word_gen(model, t, curr_word, num):
          #모델,토큰나이저,입력단어,개수
        ini_word=curr_word
        sent=''
        for _ in range(num):
            #print(t.word_index)
            encoded=t.texts_to_sequences([curr_word])[0] #[3]
            #print(encoded)
            encoded=pad_sequences([encoded],maxlen=23,padding='pre') # [[0 0 0 0 3]]
            #print(encoded)           
            res=model.predict_classes(encoded)
            #print(res) #4
            for word, index in t.word_index.items():
                if res==index: #동일한 단어를 찾았다면
                    break
            curr_word=curr_word+' '+word 
            sent=sent+' '+word            
        sent=ini_word+sent
        return sent      
print(word_gen(model, t, 'saturday', 10))    
#멀티캠퍼스에 라는 단어 뒤에 3개의 단어 예측
```
