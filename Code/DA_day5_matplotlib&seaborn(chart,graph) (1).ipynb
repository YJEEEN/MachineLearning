### xlabel vs xticks
label은 축에 대한 이름 
ticks는 각 bar에 대한 이름을 지정

```python
y=[3,1,2]
x = np.arange(len(y))
plt.xlabel('ABC')
plt.ylabel('Freq')
xlabel=['X','Y','Z']
plt.xticks(x,xlabel)
plt.bar(x,y)
```

```python
data=['A','B','C','D']
y_pos=np.arange(len(data))
perform=2+3*np.random.rand(len(data))
perform
plt.title('My chart')
plt.xlabel('xlabel')
plt.barh(y_pos,perform,alpha=0.7, xerr=0.2) 
# barh는 bar horizontal --> 가로로 출력
# alpha=1 : 투명도 1은 default 0.7은 30%의 불투명도
# xerr : 0.2라는 오차범위를 표시
plt.yticks(y_pos,data) #'A'는 y_pos를 시작 위치로 출력됨
plt.show()
```

### stem plot : 폭이 없는 막대bar (이산확률 그래프 같은 것 표현시 사용)

```python
x = np.linspace(0.1,2*np.pi,10) # 선형공간 함수 0.1~ 6.28까지를 10구간으로 나눔
np.cos(x) 
plt.stem(x,np.cos(x), '-.') # 선 모양에 대한 옵션 '-':dafault/ '-.' : 점선으로 표현 --> API 참고할 것!!
```

### pie graph (카테고리별 비교시 사용)

```python
labels = ['Seoul','Busan','Jejy','Kwangju']
colors = ['gold','silver','blue', 'lightskyblue']
sizes = [15, 30, 45, 10] # 합이 100(%)
explode=(0,0,0.1,0)
plt.title('Pie chart')
plt.pie(sizes, labels=labels, 
        colors=colors, explode=explode, startangle=90, shadow=True,
        autopct='%1.1f%%') 
# shadow=Ture --> '성의' 넣어줘라...
# colors를 지정해서 대입할 수도 있다
# startangle:12시 반시계 방향으로 데이터순서대로 출력
# explode : 원래 파이차트와의 gap 
# autipct : %는무조건 출력해줘야하고, 뒤에 %%는 '%'를 출력용 1.1f는 1자릿수확보후, 소숫점1번째자리까지 출력
```

### imshow (2차원 이미지 데이터를 색상으로 표시)

```python
from sklearn.datasets import load_digits # 수치로 저장된 이미지들을 호출
digits = load_digits()  # load_digits()하면 불러와짐
digits # dic형식으로 저장되어있고, 5개의 key 값에 이미지가 저장되어 있음
x = digits.images[0] 
# 8*8 이미지인데, 0값은 흰색 / 숫자가 클 수록 진한 색
# digits.images[n]에는 숫자 이미지가 array로 저장되어 있다
plt.imshow(x,cmap=plt.cm.bone_r, interpolation='nearest')
# cmap=plt.cm.bone_r 다양한 컬러 제공됨
# born_r :grayscale로 보여줘서 대부분 많이 사용
# interpolation option : 색상을 자연스럽게 변하도록 'nearest' 줄 수 있다
```

### contourPlot
#### 컨투어는 단순히 등고선만 표현 컨투어f는 조금 더 색상에 주력 --> 혼용해서 사용
#### (지도 형태 --> x,y 입력 후 z 출력)
#### 컨투어플롯/ 서피스플롯

```python
def myFunc(x,y):
    return(1-x/2+x**4+y**2)*np.exp(-x**2-y**2) # 4차 방정식 * exp(자연상수2) => 임의의 수식

x = np.linspace(-3,3,256)
y = np.linspace(-3,3,256)
xx,yy = np.meshgrid(x,y)
# meshgrid()는 등고선을 그리고자 하는 대상 데이터를 나타내는 특수한 행렬이다.
zz = myFunc(xx,yy)
zz
```

```python
# plt.contourf(xx,yy,zz,alpha=0.9)
plt.contour(xx,yy,zz)
```

### heatmap(data) *중요*
#### * '열 지도'라고 한다
#### * 상관계수 --> 모든 변수간의 상관 관계를 나타낼 수 있다 (반쪽만 나타내는 mask를 많이 사용)

```python
titanic.columns
```

```python
titanic_size = titanic.pivot_table(
index='class',
columns='sex',
aggfunc = "size")  # aggfunc : 적용하려는 함수 기본 default값은 mean()을 반환한다

sns.heatmap(titanic_size, annot=True, fmt='d', cmap=sns.light_palette('red'))
# annot :  an notice
# fmt = 'd' : 포멧 정수
# cmap=sns.옵션명('컬러명') 
```
